```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as co
%matplotlib inline
```

```python
datos = np.loadtxt("IRCF.txt").T
print(datos[0])
```

```python
plt.scatter(datos[0],datos[1])
plt.errorbar(datos[0],datos[1])
plt.grid()
```

```python
h = co.h
k = co.k
c = co.c
e = 2.7182
datos[0]=datos[0]*3e10
```

```python
def I(v,T):
    I=((h*v**3)/c**2)*(1/(e**(h*v/k*T)-1))
    I=I/10e-26
    return I
```

```python
plt.scatter(datos[0],I(datos[0],5))
```
