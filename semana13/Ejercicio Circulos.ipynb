```python
#Primero se declaran las librerias que se van a utilizar
import matplotlib.pyplot as pit
import numpy as np
%matplotlib inline
```

```python
#El ejercicio consiste en tomar puntos random y tomar estos como centro de un circulo de otros puntos

#Con un primer ciclo se inicializa un punto aleatorio
for i in range(10):
    x = np.random.rand(1) #Se inicia primero en la coordenada x
    y = np.random.rand(1) #Se inicia despues en la coordenada y
    pit.scatter(x,y, color='k') #Se imprime el primer centro en color negro
    
    #Despues, para trazar los puntos alrededor del centro se tienen que calcular los puntos alrededor de este
    #Se utilizan coordenadas polares, usando un valor fijo como radio y utilizando el contador k para variar el angulo
    #Se utilizan 50 puntos alrededor de cada centro
    for k in range(50):
        a=0.1*np.cos(k) #Se calcula la coordenada polar en x
        b=0.1*np.sin(k) #Se calcula la coordenada polar en y
        pit.scatter(a+x,b+y,color='r') #Se imprimen los valores sumando (a, b) al valor del centro original (x,y)
        
#Siempre que se ejecute el programa se reiniciaran los valores de los centros y estos cambiaran de posicion
```
