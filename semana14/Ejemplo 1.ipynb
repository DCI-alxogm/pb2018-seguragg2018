```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(-20,20,100)
y=np.cos(x)
x2=np.linspace(-20,20,100)
y2=np.sin(x2)
```

```python
plt.plot(x,y)
plt.plot(x2,y2)
```

```python
fx=lambda x,y:(x**2+y**2)-25
```

```python
fx(5,5)
```

```python
plt.plot(x,fx(x,y))
```

```python
def fx2(x,y):
    z=x+y
    return z
```

```python
plt.plot(x,fx2(x,y))
```

```python
def recta(a,m=1,b=0,label='x'):
    z = m*a+b
    plt.plot(a,z,label=label)
    
```

```python
x = np.linspace(-5,5,10)
recta(x)
recta(x,-2,1,'x+1')
plt.legend()

plt.grid()
```
